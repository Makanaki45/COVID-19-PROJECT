```python
#IMPORTING NECESSARY LIBRARIES
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
import seaborn as sns
from datetime import datetime
```

```python
covid = pd.read_csv('covid_19_india.csv')
```

```python
#DROPPING UNNECESSARY COLUMNS
covid.drop(columns=['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Time'], inplace=True)
```

```python
#WE ARE CHANGING THE DATATYPE OF DATE COLUMN
covid['Date'] = pd.to_datetime(covid['Date'], format='%Y-%m-%d')
```

```python
covid.head()
```

```python
#CREATING A NEW COLUMN FOR ACTIVE CASES
covid['Active_Cases'] = covid['Confirmed'] - covid['Cured'] + covid['Deaths']
```

```python
covid.tail()
```

```python
#TRANSFORMING AND SUMMARIZING DATA USING PIVOT_TABLE
statewise = pd.pivot_table(covid, values=['Cured', 'Deaths', 'Confirmed'], index='State/UnionTerritory', aggfunc=max)
```

```python
#CREATING A NEW COLUMN FOR RECOVERY RATE
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
#CREATING A NEW COLUMN FOR MORTALITY RATE
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
#SORTING CONFIRMED IN A DESCENDING WAY
statewise = statewise.sort_values('Confirmed', ascending=False)
```

```python
#ADDING A BACKGROUND COLOR USING MATPLOTLIB COLORMAP
statewise.style.background_gradient('cubehelix')
```

```python
#TOP 10 STATES WITH HIGHEST ACTIVE CASES
top_10_active_cases = covid.groupby('State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values('Active_Cases', ascending=False).reset_index()
pt.figure(figsize=(18,7))
pt.title('Top 10 state with highest active cases')
sns.barplot(data = top_10_active_cases.iloc[:10], x='State/UnionTerritory', y='Active_Cases', linewidth=2, edgecolor='red')
pt.xlabel('state')
pt.ylabel('Total active cases')
pt.show()
```

```python
#TOP 10 STATES WITH HIGHEST DEATHS CASES
top_10_Deaths_cases = covid.groupby('State/UnionTerritory').max()[['Deaths', 'Date']].sort_values('Deaths', ascending=False).reset_index()
pt.figure(figsize=(18,5))
pt.title('Top 10 state with highest deaths cases')
sns.barplot(data = top_10_Deaths_cases.iloc[:12], x='State/UnionTerritory', y='Deaths', linewidth=2, edgecolor='red')
pt.xlabel('state')
pt.ylabel('Total Deaths cases')
pt.show()
```

```python
#GROWTH TREND FOR STATE/UNIONTERRITORY
fig = pt.figure(figsize=(18,5))
ax = sns.lineplot(data = covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x = 'Date', y = 'Active_Cases', hue= 'State/UnionTerritory')
ax.set_title('Top 5 Affected states in india')
```
