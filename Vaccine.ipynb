```python
#IMPORTING NECESSARY LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as pt
```

```python
vaccine_df= pd.read_csv('covid_vaccine_statewise.csv')
```

```python
#RENAMING A COLUMN
vaccine_df.rename(columns={'Updated On':'Vaccine_Date'}, inplace=True) 
```

```python
#CHECKING THE NULL VALUES
vaccine_df.isnull().sum()
```

```python
#FILLING NAN VALUES WITH ZERO
vaccine_df = vaccine_df.fillna(0)
```

```python
vaccine_df.head()
```

```python
#DROP SOME USEFUL cOLUMNS
vaccination = vaccine_df.drop(columns= ['Sputnik V (Doses Administered)', 'AEFI', '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'], axis=1)
```

```python
vaccination.head()
```

```python
# MALE VS FEMALE VACCINATION
male = vaccination['Male(Individuals Vaccinated)'].sum()
female = vaccination['Female(Individuals Vaccinated)'].sum()
labels = ['Male','Female']
x = np.array([male,female])
pt.pie(x, labels=labels, autopct='%0.1f%%')
pt.title('Male and Female Vaccination')
pt.legend()
pt.show()
```

```python
#FILTERING ROWS WHERE STATE = INDIA
vaccine = vaccine_df[vaccine_df['State']!='India']
vaccine.head()
```

```python
#RENAMING COLUMN TOTAL INDIVIDUALS VACCINATED TO TOTAL
vaccine.rename(columns = {'Total Individuals Vaccinated':'Total'}, inplace =True)
vaccine.head()
```

```python
#MOST VACCINATED STATE
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False)[:5]
max_vac
```

```python
#TOP 5 VACCINATED STATES IN INDIA
pt.figure(figsize=(10,5))
pt.title('Top 5 Vaccinated States in India')
ax = sns.barplot(data = max_vac.iloc[:5], y=max_vac.Total, x=max_vac.index, linewidth=2)
pt.xlabel('States')
pt.ylabel('Vaccination')
pt.show()
```
